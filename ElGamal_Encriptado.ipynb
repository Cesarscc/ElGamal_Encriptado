```python
from elgamal.elgamal import Elgamal

# Mensaje a Cifrar

m = b'Seguridad Informatica'
print("Mensaje a cifrar: ", m)

# Se genera las llaves
# pb = llave publica
# pv = llave privada
# En este caso ambas llaves son de 128 Bits

pb, pv = Elgamal.newkeys(128)
print("\n Numero primo: ", pb.p)
print("\n Generador: ", pb.g)
print("\n Llave publica: ", pb)
print("\n Llave privada: ", pv)

# Se cifra el mensaje utilizando la llave publica
ct = Elgamal.encrypt(m, pb)
print("\n Mensaje encriptado: ", ct)

dd = Elgamal.decrypt(ct, pv)
print("\n Mensaje Descifrado: ", dd)
```

```python
import random
from math import pow

a = random.randint(2,10)
#maximo comun divisor
def gcd(a,b):
  if (a<b):
    return gcd(b,a)
  elif (a%b)==0:
    return b
  else:
    return gcd(b,a%b)
  
#generador de numeros grandes
def gen_key(q):
  key = random.randint(pow(10,20),q)
  while gcd(q,key)!=1:
    key = random.randint(pow(10,20),q)
  return key

  #exponente modular
def power(a,b,c):
    x=1
    y=a
    while (b>0):
      if (b&2)==0:
        x = (x*y)%c
      y = (y*y)%c
      b = int(b/2)
    return x%c

# Encriptacion asimetrica
def encrypt(msg,q,h,g):
  en_msg = []

  k = gen_key(q)#Llave privada del que envia el mensaje cifrado
  s = power(h,k,q)
  p = power(g,k,q)

  for i in range(0,len(msg)):
    en_msg.append(msg[i])
  for i in range(0,len(en_msg)):
    en_msg[i] = s*ord(en_msg[i])
  return en_msg, p

def decrypt(en_msg, p, key, q):
  dr_msg = []
  h = power(p,key,q)
  for i in range(0, len(en_msg)):
    dr_msg.append(chr(int(en_msg[i]/h)))
  return dr_msg

#main
msg = 'Seguridad Informatica'
print("Mensaje Original: ", msg)

q = random.randint(pow(10,20),pow(10,50))#Numero grande
g = random.randint(2,q)#Numero usado dentro del numero grande

key = gen_key(q) #Llave privada para el receptor
h = power(g,key,q)#Llave publica

print("Numero aleatorio q: ", q)
print("\n Generador g: ",g)
print("\n Llave privada : ",key)

print("\n h=g^a: ",h)

en_msg, p = encrypt(msg,q,h,g)#Mensaje Encriptado

print("\n Mensaje Cifrado: ",en_msg)

dr_msg = decrypt(en_msg,p,key,q)#Mensaje Desencriptado
dmsg = ''.join(dr_msg)
print("\n Mensaje descifrado: ", dmsg)
```
